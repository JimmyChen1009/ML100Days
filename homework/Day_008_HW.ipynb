```python
app_train['AMT_INCOME_TOTAL'].describe()
```

```python
t=app_train['AMT_INCOME_TOTAL'].max()
print(f'Max= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].min()
print(f'Max= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].mean()
print(f'Mean= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].median()
print(f'Median= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].var()
print(f'Variance= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].std()
print(f'Standard Deviation= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].skew()
print(f'Skewness= {t}\n')
t=app_train['AMT_INCOME_TOTAL'].kurt()
print(f'Kurtosis= {t}\n')
```

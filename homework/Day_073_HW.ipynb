# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
import matplotlib.pyplot as plt
import numpy as np

precision = 1e-6 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
```

```python
# 需要執行初始值設定, 下列三個
# Learning rate
# iteration counter
# Step size  

lr_list = [0.0001, 0.001, 0.01, 0.1, 1, 10]
iters_count_list = []
for lr in lr_list:
    cur_x = 3 # The algorithm starts at x=3
    iters = 0
    step_size = 1
    while step_size > precision and iters < max_iters:
        prev_x = cur_x
        cur_x = cur_x - lr * df(prev_x)
        step_size = abs(cur_x - prev_x)
        iters = iters + 1
    iters_count_list.append(iters)

plt.plot(iters_count_list)
plt.xticks(np.arange(0,6), lr_list, rotation='vertical')
plt.xlabel("learning rate")
plt.ylabel("iterations")
```

- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_name = iris.feature_names
label_name = iris.target_names

print("Shape", X.shape)
print("feature name: ", feature_list)
print("label name: ", label_name)
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

### Use 2 Features
- sepal length vs sepal width
- petal length vs petal width

```python
# sepal
x = X[:,0:2]

km = KMeans(n_clusters=3, random_state=42)
km.fit(x)
y_pred = km.predict(x)
```

```python
colors = ['b', 'c', 'y']

plt.figure(figsize=(8, 6)) 
plt.subplot(1, 2, 1)
plt.scatter(x[:,0], x[:,1],c=y)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("True")
plt.subplot(1, 2, 2)
plt.scatter(x[:,0], x[:,1],c=y_pred)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("K-Means")
plt.tight_layout()
```

```python
# petal
x = X[:,2:4]

km = KMeans(n_clusters=3, random_state=42)
km.fit(x)
y_pred = km.predict(x)
```

```python
plt.figure(figsize=(8, 6)) 
plt.subplot(1, 2, 1)
plt.scatter(x[:,0], x[:,1],c=y)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("True")
plt.subplot(1, 2, 2)
plt.scatter(x[:,0], x[:,1],c=y_pred)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("K-Means")
plt.tight_layout()
```

### Different K

```python
x = X[:,2:4]
for i in range(1, 5):
    km = KMeans(n_clusters=i*2, random_state=42)
    km.fit(x)
    y_pred = km.predict(x)
    plt.subplot(2,2,i)
    plt.scatter(x[:,0], x[:,1], c=y_pred)
    plt.title("K= "+ str(i*2))
    plt.tight_layout()
```

### Different Initialisation

```python
km = KMeans(n_clusters=3, init="k-means++")
km.fit(X)
y_pred = km.predict(X)
plt.scatter(x[:,0], x[:,1], c=y_pred)
plt.title("k-means init")
plt.tight_layout()
```

```python
km = KMeans(n_clusters=3, init="random", n_init=1)
km.fit(X)
y_pred = km.predict(X)
plt.scatter(x[:,0], x[:,1], c=y_pred)
plt.title("random initialisation")
plt.tight_layout()
```

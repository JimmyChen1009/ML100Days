# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def relu(x):
    return np.maximum(0, x)

def reluDerivative(x):
    return (x > 0) * 1
```

```python
x = plt.linspace(-3,3,100)
plt.subplot(2,1,1)
plt.plot(x, relu(x))
plt.title("ReLU")
plt.subplot(2,1,2)
plt.plot(x, reluDerivative(x))
plt.title("DReLU")
plt.tight_layout()
```
